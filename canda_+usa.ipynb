```python
import pandas as pd

# Load the dataset
df_usa = pd.read_csv("USA_dataset.csv")

if 'Area' in df_usa.columns and 'CCME_Values' in df_usa.columns:

    # Step 1: Group by Area to compute mean, median, std
    stats = df_usa.groupby('Area')['CCME_Values'].agg(['mean', 'median', 'std'])

    # Step 2: Add a count column (number of rows per Area)
    stats['count'] = df_usa['Area'].value_counts()

    # Step 3: Reset index for filtering
    stats = stats.reset_index()

    # Step 4: Apply filter
    filtered_stats = stats[
        (stats['median'] >= 93) & (stats['median'] <= 97) &
        (stats['std'] >= 5) & (stats['std'] <= 10)
    ]

    # Step 5: Display the result
    print("Filtered Areas (Median ≈ 95, Std 5–10) with Data Count:")
    print(filtered_stats[['Area', 'mean', 'median', 'std', 'count']])

else:
    print("Missing 'Area' or 'CCME_Values' column.")
```

```python
# Step: Compute counts per area
counts = df_usa['Area'].value_counts()

# Filter areas with >= 1000 samples
areas_50_plus = counts[counts >= 50].index.tolist()

# Filter dataset for these areas
df_large_areas = df_usa[df_usa['Area'].isin(areas_50_plus)]

# Then compute stats and filter based on median and std
stats = df_large_areas.groupby('Area')['CCME_Values'].agg(['mean', 'median', 'std']).reset_index()

filtered_stats = stats[
    (stats['median'] >= 93) & (stats['median'] <= 97) &
    (stats['std'] >= 5) & (stats['std'] <= 10)
]

print(filtered_stats)
```
